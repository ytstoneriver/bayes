```python
# 環境変数からGitHubのパーソナルアクセストークンを取得
github_token = os.getenv("GITHUB_TOKEN")

# .netrcファイルを作成してGitHubの認証情報を設定
netrc_path = os.path.expanduser("~/.netrc")
with open(netrc_path, "w") as file:
    file.write(f"machine github.com\nlogin {github_token}\npassword x-oauth-basic\n")

# .netrcファイルのパーミッションを設定
os.chmod(netrc_path, 0o600)
```

```python
import os
import subprocess

# Google Driveをマウント
from google.colab import drive
drive.mount('/content/drive')

# プロジェクトディレクトリを定義
root_dir = "/content/drive/MyDrive/Colab Notebooks/github"
sample_dir = f"{root_dir}/sample"

# サブモジュールを削除
subprocess.run(f'cd "{root_dir}" && git submodule deinit -f {sample_dir}', shell=True)
subprocess.run(f'cd "{root_dir}" && git rm -f {sample_dir}', shell=True)

# .gitmodulesファイルからサブモジュールのエントリを削除
subprocess.run(f'cd "{root_dir}" && git config -f .gitmodules --remove-section submodule.sample', shell=True)

# キャッシュをクリア
subprocess.run(f'cd "{root_dir}" && git config --remove-section submodule.sample', shell=True)

# 変更をコミット
subprocess.run(f'cd "{root_dir}" && git add .', shell=True)
subprocess.run(f'cd "{root_dir}" && git commit -m "Remove sample submodule"', shell=True)

# リモートリポジトリにプッシュ
subprocess.run(f'cd "{root_dir}" && git push origin main', shell=True)
```
